```python
# Import necessary libraries
import pandas as pd

# Load the Excel file with the correct path
file_path = '/content/sample_data/data_dummy_p.xlsx'  # Path yang sesuai setelah diupload
xls = pd.ExcelFile(file_path)

# Load the data from the first sheet
df = pd.read_excel(xls, sheet_name='Sheet1')

# Melanjutkan proses cleansing seperti sebelumnya
df['Tanggal'] = pd.to_datetime(df['Tanggal'], errors='coerce')
df['PT'] = df['PT'].str.strip()
df['Bank'] = df['Bank'].str.strip().str.upper()  # Pastikan nama bank konsisten (gunakan upper case)
df['Jenis Transaksi'] = df['Jenis Transaksi'].str.strip()

df['Pemasukkan (Rp)'] = pd.to_numeric(df['Pemasukkan (Rp)'], errors='coerce')
df['Pengeluaran (Rp)'] = pd.to_numeric(df['Pengeluaran (Rp)'], errors='coerce')
df['Saldo Sisa (Rp)'] = pd.to_numeric(df['Saldo Sisa (Rp)'], errors='coerce')

# Mengatasi nilai yang hilang (NaN)
df = df.fillna({'Pemasukkan (Rp)': 0, 'Pengeluaran (Rp)': 0, 'Saldo Sisa (Rp)': 0})

# Menampilkan data yang telah dibersihkan
df.head()
```

```python
import pandas as pd
from datetime import datetime

def add_daily_data(existing_data, new_data):
    """
    Fungsi ini akan menambahkan data baru ke dataset yang sudah ada (existing_data),
    dan memastikan data yang ditambahkan memiliki tanggal yang lebih baru.

    existing_data: DataFrame, dataset yang sudah ada.
    new_data: DataFrame, dataset dengan data baru yang akan ditambahkan.
    """

    # Pastikan format kolom 'Tanggal' pada data baru dan yang sudah ada konsisten
    new_data['Tanggal'] = pd.to_datetime(new_data['Tanggal'], errors='coerce')
    existing_data['Tanggal'] = pd.to_datetime(existing_data['Tanggal'], errors='coerce')

    # Menambahkan data baru jika tanggal baru lebih besar dari tanggal data yang ada
    latest_date_existing = existing_data['Tanggal'].max()
    new_data_filtered = new_data[new_data['Tanggal'] > latest_date_existing]

    # Gabungkan data baru yang sudah difilter dengan data yang sudah ada
    updated_data = pd.concat([existing_data, new_data_filtered], ignore_index=True)

    # Mengembalikan data yang telah diperbarui
    return updated_data

# Contoh penggunaan
existing_data = df  # Dataset yang sudah ada
new_data = pd.read_excel('/content/sample_data/new_data_dummy_p.xlsx')  # Data baru yang diupload

# Menambahkan data baru ke dataset yang sudah ada
updated_data = add_daily_data(existing_data, new_data)

# Tampilkan data yang telah diperbarui
updated_data.head()

# Simpan hasil gabungan ke file Excel baru
output_path = '/content/sample_data/updated_data.xlsx'
updated_data.to_excel(output_path, index=False)

# Berikan opsi untuk mengunduh file yang telah dibuat
files.download(output_path)
```

```python
def calculate_monthly_expenditure(df, start_date=None, end_date=None):
    """
    Fungsi ini menghitung total pengeluaran per bulan dari masing-masing PT dan Bank.

    df: DataFrame, dataset yang berisi data transaksi.
    start_date: string, tanggal mulai dalam format 'YYYY-MM-DD' (optional).
    end_date: string, tanggal akhir dalam format 'YYYY-MM-DD' (optional).

    Returns:
        DataFrame dengan total pengeluaran per bulan untuk masing-masing PT dan Bank.
    """

    # Ekstrak bulan dan tahun dari kolom 'Tanggal'
    df['Year-Month'] = df['Tanggal'].dt.to_period('M')

    # Jika start_date dan end_date diberikan, filter berdasarkan rentang tanggal
    if start_date and end_date:
        df = df[(df['Tanggal'] >= start_date) & (df['Tanggal'] <= end_date)]

    # Kelompokkan berdasarkan 'Year-Month', 'PT', dan 'Bank', kemudian hitung total pengeluaran
    monthly_expenditure = df.groupby(['Year-Month', 'PT', 'Bank'])['Pengeluaran (Rp)'].sum().reset_index()

    # Tampilkan hasil total pengeluaran per bulan per PT dan Bank
    return monthly_expenditure

# Contoh penggunaan:
# Misalnya kita ingin melihat pengeluaran bulan Januari 2025
start_date = '2025-02-01'
end_date = '2025-02-28'
monthly_expenditure_jan_2025 = calculate_monthly_expenditure(df, start_date=start_date, end_date=end_date)

# Tampilkan hasil pengeluaran Januari 2025
monthly_expenditure_jan_2025.head()

# fungsi tanpa filter tanggal:
# monthly_expenditure_all = calculate_monthly_expenditure(df)

# Tampilkan hasil semua pengeluaran
# monthly_expenditure_all.head()
```
